```python
import pandas as pd
d1 = {'Key1' : [1,2,3,4,5], 
      'key2' : [6,7,8,9,10],
      'key3' : [11,12,13,14,15],
      'key4' : [16,17,18,19,20]
}
df1 = pd.DataFrame(d1)
```

Here Keys will be as column and values in rows. /n
Number of elements in  every key should be same. otherwise will get the value Error.

```python
pd.DataFrame(d1, index=['a', 'b', 'c', 'd', 'e'])
```

```python
d2 = {'Key1' : [1,2,3,4,"pvp"], 
      'key2' : [6,7,"pawar",9,10],
      'key3' : [11,12,13,14,"FSDS"],
      'key4' : [16,17,18,"pw","python"]
}
df2 = pd.DataFrame(d2)
df2
```

```python
pd.merge(df1,df2)
```

By default, merge option do inner operation. Here inner is same as SQL inner i.e. intersection of two dataset.

```python
pd.merge(df1,df2, how='inner', on='Key1')
```

```python
pd.merge(df1,df2, how='inner', on='key4')
```

```python
pd.merge(df1,df2, how='left', on='key4')
```

```python
pd.merge(df1,df2, how='right', on='key4')
```

```python
pd.merge(df1,df2, how='outer', on='key4')
```

```python
pd.merge(df1,df2, how='outer', on=['key4', 'Key1'])
```

```python
d10 = {'key1' : [1,2,3,4,5], 
      'key2' : [6,7,8,9,10],
      'key3' : [11,12,13,14,15],
      'key4' : [16,17,18,19,20]
}
df10 = pd.DataFrame(d10, index =[1,2,3,4,"pvp"])
d20 = {'key11' : [1,2,3,4,"pvp"], 
      'key12' : [6,7,"pawar",9,10],
      'key13' : [11,12,13,14,"FSDS"],
      'key14' : [16,17,18,"pw","python"]
}
df20 = pd.DataFrame(d20, index =[1,2,3,4,5])
```

```python
df10
```

```python
df10.join(df20, how = 'right')
```

```python
df10.join(df20, how = 'outer')
```

```python
df10.join(df20, how = 'inner')
```

```python
## Difference between merge and join 
# Join checks the indexes whereas merge checks for the values.
# if column indexes are same then we can't perform join but merge operation allowed.
# meaning of left, right, inner and outer are same in both the cases.
```
